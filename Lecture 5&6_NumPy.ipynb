+ for numpy arrays a = a+1 adds 1 to all array elements, no loop required plus a lot faster than iterating with loop.

```python
print(id(a))
a = a+1
print(id(a))
print(a)
a = a*a
print(a)
```

## Broadcasting

+ In case of matrices if dimensions are not the same operations cannot be performed. However due to broadcasting if one of the dimensions is one and the other is the same we will still be able to perform all operations.

+ In case of compatibility the matrix is stretched along the dimension with 1 to match the dimensions of the other matrix

```python
a = np.random.randint(1,10,(1,3))
b = np.random.randint(1,10,(5,1))
c = a+b
print(a)
print(b)
print(c)
```
