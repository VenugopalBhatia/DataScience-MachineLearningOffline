+ Seaborn used for plotting, uses matplotlib internally.
+ Better version interms of visual, better colours,etc.
+ Works directly on dataframe while matplotlib doesn't.
+ for eg: For Matplotlib and Dataframe data = plt.plot(data['age],data['Salary']), For Seaborn= fn(x = 'data',y = 'Salary',data = df_)
+ sns.get_dataset_names to get internal sample datasets.
+ sns.load_dataset to load dataset.
+ Uses plt.show() to see plot
+ sns.scatterplot for scatter plots.
+ sns.relplot() : By default creates scatter plot, can also create lineplots, mention kind = 'line'.
> + Refer to documentation for parameters.
> + hue categorises based on colors, style on marker style.
> + size for markers' sizes. size takes as per category, sizes can explicitly change the kind of size it will have. We can scale the sizes to the range mentioned.
> + This is a wrapper for line and scatterplot.
+ Catplot is a wrapper for a wide range of plots.
+ Catplot is used to create plots between categorical and numerical values.
> + jitter is used to change overlap between data points.
> + swarm swarm by default does separation of datapoints. Zero overlapping is observed.
+ Swarm used when data points are less to observe each datapoint without overlap.
+ Distribution plots to observe how values are distributed as according to category.
+ In a boxplot the lower line indicates the 25th percentile, the middle line indicates the 50th percentile and the upper most line is the 75th percentile, the lines show data above and below the box and the pointsindicate outliers.
+ We use lmplot to visualize regression models on data

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
sns.get_dataset_names
tips = sns.load_dataset('tips')
```

```python
x = tips.total_bill
y = tips.tip
sns.scatterplot(x = x,y = y,data = tips)
plt.show()
```

```python
sns.relplot(x = 'total_bill',y = 'tip',data = tips,kind = 'line')
plt.show()
```

```python
sns.relplot(x = 'total_bill',y = 'tip',data = tips,kind = 'scatter',hue = 'smoker',style = 'sex')
plt.show()
```

```python
sns.relplot(x = 'total_bill',y = 'tip',data = tips,kind = 'scatter',hue = 'smoker',style = 'sex',palette = ['b','g'],size='size' )
plt.show()
```

```python
sns.relplot(x = 'total_bill',y = 'tip',data = tips,kind = 'scatter',hue = 'smoker',palette = ['g','b'],style = 'sex',size='size',sizes = (100,600))
plt.show()
```

```python
sns.catplot(x = 'day',y = 'total_bill',data = tips,jitter = 0.3)
```

```python
sns.catplot(x = 'day',y = 'total_bill',data = tips,kind = 'swarm',hue = 'smoker')
```

```python
sns.catplot(x = 'time',y = 'tip',data = tips,kind = 'swarm',hue = 'sex',dodge = False)
```

```python
sns.catplot(x = 'day',y = 'total_bill',data = tips,kind = 'boxen',dodge = True)
```

```python
sns.catplot(x = 'day',y = 'total_bill',data = tips,kind = 'violin')
plt.grid()
```

```python
sns.catplot(x = 'size',data = tips,kind = 'count')
```

```python
sns.lmplot(x = 'total_bill',y = 'tip',data = tips,hue = 'sex',order = 2)
```
