```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_1 = pd.read_csv(r'ShareRaceByCity.csv')
```

```python
df_1
```

```python
df_1.replace('(X)',0,inplace = True)
race = df_1.columns[2:]
for i in race:
    df_1[i] = df_1[i].astype(float)
```

```python
G = np.unique(df_1['Geographic area'])
```

```python
M = {}
Mw = df_1.groupby('Geographic area').share_white.mean()
MBl = df_1.groupby('Geographic area').share_black.mean()
Mnam = df_1.groupby('Geographic area').share_native_american.mean()
Mas = df_1.groupby('Geographic area').share_asian.mean()
Mhis = df_1.groupby('Geographic area').share_hispanic.mean()
M['Geographic area'] = Mw.index
M['share_white'] = Mw.values
M['share_black'] = MBl.values
M['share_native_american'] = Mnam.values
M['share_asian'] = Mas.values
M['share_hispanic'] = Mhis.values
```

```python
M = pd.DataFrame(M)
M
```

```python
plt.figure(figsize = (15,10))
sns.barplot(x = M['Geographic area'],y = M['share_black'],data = M,color = 'blue')
sns.barplot(x = M['Geographic area'],y = M['share_white'],data = M,color = 'red')
sns.barplot(x = M['Geographic area'],y = M['share_hispanic'],data = M,color= 'orange')
sns.barplot(x = M['Geographic area'],y = M['share_asian'],data = M,color = 'black')
sns.barplot(x = M['Geographic area'],y = M['share_native_american'],data = M,color = 'yellow')
plt.ylabel("Percentage of State's Population According to Races ")
plt.show()
```
